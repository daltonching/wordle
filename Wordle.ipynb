```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Open and read contents of file
with open('C:\\Users\\Dalton\\Downloads\\WordleLetterDistribution-main\\WordleLetterDistribution-main\\WordleAnswers.txt') as src:
    og = src.read()
    
# strip double quotes
    og = og.replace('"', '')
    
# make solution set list
    words = [i for i in og[1:-1].split(',')]
```

```python
# Split each word into individual letters
def test(wordz):
    letterz = []
    for i in wordz:
        letterz.append(list(i))
    return letterz
testlist = test(words)
print(testlist)
```

```python
# Count number of specific letter in a word
def num_of(a,testinglist):
    total = 0
    for i in testinglist:
        for x in i:
            if x == a:
                total += 1
    return total
```

```python
# Count number of specific letters of the first letter place
def num_of1(a,testinglist):
    total = 0
    for i in testinglist:
        for x in i[0]:
            if x == a:
                total += 1
    return total
```

```python
# Count number of specific letter of the second letter place
def num_of2(a,testinglist):
    total = 0
    for i in testinglist:
        for x in i[1]:
            if x == a:
                total += 1
    return total
```

```python
#Find frequency of letter at each position to put into a dict
# In other words extracting data of each letter at all places in the word
attempt = {}
def library1(a,testinglist):
    attempt = {}
    total = 0
    for i in testinglist:
        for x in i[0]:
            if x == a:
                total += 1
                attempt['1st'] = total
    total = 0
    for i in testinglist:
        for x in i[1]:
            if x == a:
                total += 1
                attempt['2nd'] = total
    total = 0
    for i in testinglist:
        for x in i[2]:
            if x == a:
                total += 1
                attempt['3rd'] = total
    total = 0
    for i in testinglist:
        for x in i[3]:
            if x == a:
                total += 1
                attempt['4th'] = total
    total = 0
    for i in testinglist:
        for x in i[4]:
            if x == a:
                total += 1
                attempt['5th'] = total
    return attempt
```

```python
# Test to see if function is working properly
library1('s',testlist)
```

```python
# Create a masterlist for our data
masterlist = {}
for i in range(97,123):
    masterlist[chr(i).format(i).upper()] = library1(chr(i).format(i), testlist)
print(masterlist)
```

```python
# Convert masterlist into a dataframe
df = pd.DataFrame(masterlist)
print(df)
```

```python
# plot the data
ax = df.plot(kind='bar', legend = True, figsize=(18,15), width = .85)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```
